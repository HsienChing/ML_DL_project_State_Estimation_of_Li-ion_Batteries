### 2.1. A bug in the example code of "Keras layers" subsection

**Position:**  
"Keras layers" subsection

**Link:**  
https://www.tensorflow.org/guide/intro_to_modules#keras_layers

**Condition:**  
There is a bug in the example code of "Keras layers" subsection. I didn't see the expected message given by the tutorial.

I have extracted the example and left the error message during execution.

I have tried the "[Run in Google Colab](https://colab.research.google.com/github/tensorflow/docs/blob/master/site/en/guide/intro_to_modules.ipynb)" file from the official tutorial, "Restart session and run all", the same error message occurs.

**Suggestion:**  

**Possible solution**

I have tried the "Explain error" given by Google Colab. 

The error occurs because the Keras layer `simple_layer` expects a TensorFlow tensor as input, but it's receiving a plain Python list. To resolve this, you need to explicitly convert the input list to a `tf.constant`.

As the `tf.constant` is added, the example code can be run normally.

Example code after modification (Result: can run normally)
```python
simple_layer(tf.constant([[2.0, 2.0, 2.0]]))
```

Example code before modification (Result: an error message occurs)
```python
simple_layer([[2.0, 2.0, 2.0]])
```
