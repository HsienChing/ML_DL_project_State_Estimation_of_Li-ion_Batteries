# TensorFlow installation verfication

A brief installation guide is provided in https://github.com/HsienChing/ML_DL_project_State_Estimation_of_Li-ion_Batteries/blob/main/development_environment_setup/Windows_WSL2_Python_3.12_TensorFlow_2.19.1.md  
Or follow the official installation guide: https://www.tensorflow.org/install

After the installation of `TensorFlow`, we can use the following codes to verify the installation.

## 1. Verify the CPU setup

### 1.1. In Python virtual environment (venv)

Use the code to verify.

````python
python3 -c "import tensorflow as tf; print(tf.reduce_sum(tf.random.normal([1000, 1000])))"
````

If a tensor is returned, we've installed `TensorFlow` successfully.

### 1.2. In Jupyter notebook 

Use the code to verify.

````python
import tensorflow as tf; print(tf.reduce_sum(tf.random.normal([1000, 1000])))
````

If a tensor is returned, we've installed `TensorFlow` successfully.

## 2. Verify the CPU setup

### 2.1. In Python virtual environment (venv)

Use the code to verify.  

````python
python3 -c "import tensorflow as tf; print(tf.config.list_physical_devices('GPU'))"
````

If a list of GPU devices is returned, we've installed TensorFlow successfully.

Example message:  
Usually appears in the last line  
```Python
[PhysicalDevice(name='/physical_device:GPU:0', device_type='GPU')]
```

### 2.2. In Jupyter notebook 

Use the code to verify.

````python
import tensorflow as tf; print(tf.config.list_physical_devices('GPU'))
````

If a list of GPU devices is returned, we've installed TensorFlow successfully.

Example message:  
Usually appears in the last line  
```Python
[PhysicalDevice(name='/physical_device:GPU:0', device_type='GPU')]
```
