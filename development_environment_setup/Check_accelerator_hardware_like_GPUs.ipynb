# Check accelerator hardware like GPUs

Running large calculations of `TensorFlow` on CPU can be slow. When properly configured, `TensorFlow` can use accelerator hardware like GPUs to execute operations very quickly.

Use the code to check:

```python
import tensorflow as tf

if tf.config.list_physical_devices('GPU'):
    print("TensorFlow **IS** using the GPU")
else:
    print("TensorFlow **IS NOT** using the GPU")
```

As we see in the last line "TensorFlow \*\*IS\*\* using the GPU"
