# Linear Regression

REF: https://en.wikipedia.org/wiki/Linear_regression


In statistics, `linear regression` is a model that estimates the relationship between a scalar response (`dependent variable`) and one or more explanatory variables (`regressor` or `independent variable`).

"`Linear regression` is linear in its unknown parameters (i.e., the parameters appear only to the first power) not necessarily in variables."

From a `mathematical perspective`, $\mathbf{X}$ and $\mathbf{Y}$ are treated as variables, and the parameters are considered fixed constants but from a `statistics perspective`, the focus is on the parameters. Once we substitute observed data for $\mathbf{X}$ and $\mathbf{Y}$, the model becomes a function of the parameters, which then behave like variables that need to be estimated.

**Examples of linear regression**

1. $$ \mathbf{Y} = \beta_{0} + \beta_{1}     \mathbf{X}     + \epsilon $$
2. $$ \mathbf{Y} = \beta_{0} + \beta_{1}     \mathbf{X}^{2} + \epsilon $$
3. $$ \mathbf{Y} = \beta_{0} + \beta_{1} (ln \mathbf{X})    + \epsilon $$

In the above equations, the coefficients $\beta_{0}$ and $\beta_{1}$ (parameters) are linear (i.e., raised to the power of 1), where $\epsilon$ is the error term. 

|     |        $\mathbf{X}$  |      $\mathbf{Y}$  |
| --- | -------------------- | ------------------ |
|   1 | independent variable | dependent variable |
|   2 |                input |             output |

**Examples of non-linear regression**

1. $$ \mathbf{Y} =   \beta_{0} + \beta_{1}^{2} \mathbf{X} + \epsilon $$
2. $$ \mathbf{Y} =            (ln\beta_{1})    \mathbf{X} + \epsilon $$

In the above equations, the coefficients $ \beta_{1}^{2}$ and $ln\beta_{1}$ (parameters) are non-linear.

A model with exactly one explanatory variable is a `simple linear regression`; a model with two or more explanatory variables is a `multiple linear regression`. This term is distinct from `multivariate linear regression`, which predicts multiple correlated dependent variables rather than a single dependent variable.

## Dependent and independent variables

REF: https://en.wikipedia.org/wiki/Dependent_and_independent_variables

A variable is considered `dependent` if it depends on (or is hypothesized to depend on) an `independent variable`. 

`Dependent variables` are the outcome of the test they depend on, by some law or rule (e.g., by a mathematical function). 

`Independent variables` are not seen as depending on any other variable in the scope of the experiment in question.

### In pure mathematics 

**Example 1:**  
The most common formula $y = f(x)$.

|     | independent variable | dependent variable |
| --- | -------------------- | ------------------ |
|   1 |                  $x$ |                $y$ |
|   2 |                input |             output |

It is possible to have `multiple independent variables` or `multiple dependent variables`. 

**Example 2:**  
In multivariable calculus, one often encounters functions of the form $z = f(x, y)$, where $z$ is a dependent variable and $x$ and $y$ are independent variables.

|     | independent variable | dependent variable |
| --- | -------------------- | ------------------ |
|   1 |             $x$, $y$ |                $z$ |
|   2 |                input |             output |

**Example 3:**  
Functions with multiple outputs are often referred to as [vector-valued functions](https://en.wikipedia.org/wiki/Vector-valued_function).

In the linear case, the function $y = a x$ can be expressed in terms of matrices:
$$ \mathbf{y} = A \mathbf{x} , $$
where $\mathbf{y}$ is an ($n \times 1$) output vector, $x$ is a ($k \times 1$) vector of inputs, and $A$ is an ($n \times k$) matrix of parameters (coefficients).

$$ \mathbf{y}_{n \times 1} = A_{n \times k} \mathbf{x}_{k \times 1} $$

|     |         independent variable |           dependent variable |
| --- | ---------------------------- | ---------------------------- |
|   1 | $\mathbf{x}$ ($k$ variables) | $\mathbf{y}$ ($n$ variables) |
|   2 |                        input |                       output |

### In modeling and statistics

In mathematical modeling, the relationship between the set of `dependent variables` and set of `independent variables` is studied.

In the simple `stochastic linear model`, 
$$y_i = a + b x_i + \epsilon_i , $$ 
the term $y_i$ is the $i$-th value of the dependent variable and $x_i$ is the $i$-th value of the independent variable. The term $\epsilon_i$ is known as the "error" and contains the variability of the dependent variable not explained by the independent variable.

With multiple independent variables, the model is 
$$y_i = a + b x_{i,1} + b x_{i,2} + \dots + b x_{i,n} + \epsilon_i , $$
where $n$ is the number of independent variables.

In statistics, more specifically in linear regression, a scatter plot of data is generated with $X$ as the independent variable and $Y$ as the dependent variable. This is also called a `bivariate dataset`, $\{(x_1, y_1), (x_2, y_2), ..., (x_i, y_i)\}$. The simple linear regression model takes the form of 
$$Y_i = a + B x_i + U_i ,$$
where $i = 1, 2, ..., n$. In this case, $U_i$ are independent random variables. This occurs when the measurements do not influence each other. Through propagation of independence, the independence of $U_i$ implies independence of $Y_i$, even though each $Y_i$ has a different [expectation value](https://en.wikipedia.org/wiki/Expected_value). Each $U_i$ has an expectation value of 0 and a `variance` of $\sigma^2$ (the square of the `standard deviation`, $\sigma$). 

Proof of the expectation value of $Y_i$:

$$ \operatorname {E} [Y_{i}] = \operatorname {E} [\alpha + \beta x_{i} + U_{i}] 
= \alpha + \beta x_{i} + \operatorname {E} [U_{i}] 
= \alpha + \beta x_{i}, $$
where $\operatorname {E}$ denotes the operator to calculate expectation value.

The line of best fit for the bivariate dataset takes the form $y = \alpha + \beta x$ and is called the regression line. $\alpha$ and $\beta$ correspond to the intercept (or offset, or bias) and slope, respectively.

## Simple linear regression

REF: https://en.wikipedia.org/wiki/Simple_linear_regression

In statistics, `simple linear regression (SLR)` is a `linear regression` model with a single explanatory variable. That is, it concerns two-dimensional sample points with one independent variable and one dependent variable (conventionally, the *x* and *y* coordinates in a `Cartesian coordinate system`) and finds a linear function (a non-vertical straight line) that, as accurately as possible, predicts the dependent variable values as a function of the independent variable. The adjective *simple* refers to the fact that the outcome variable is related to a single predictor.

Consider the model function
$$ y = \alpha + \beta x , $$
which describes a line with slope $\beta$ and *y*-intercept $\alpha$. 

In general, such a relationship may not hold exactly for the largely unobserved population of values of the independent and dependent variables; we call the unobserved deviations from the above equation the errors. 

Suppose we observe $n$ data pairs and call them $\{(x_i, y_i)\}_{i = 1, 2, 3, ..., n}$. We can describe the underlying relationship between $y_i$ and $x_i$ involving this error term $\epsilon_i$ by

$$ y_{i} = \alpha + \beta x_{i} + \epsilon _{i} . $$

This relationship between the true (but unobserved) underlying parameters $\alpha$ and $\beta$ and the data points is called a `linear regression model`.
