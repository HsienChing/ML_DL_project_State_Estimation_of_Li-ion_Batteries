# Epoch, Batch, Batch Size, Weight Update, and Iteration

## 1. Definitions and descriptions

These several terms define how a machine learning model "sees" and learns from data during the training process:

**Epoch:**

A single `epoch` represents one complete pass of the entire training dataset (`input-output pairs`) through the neural network.

Models typically require many epochs to learn complex patterns and reach convergence.

Data scientists/researchers repeatedly feed the same training data to the model, so it can identify patterns and gain a deeper understanding of the new data. After every epoch of model training, the model updates its internal parameters (adjusts weights and biases) based on the data it was just fed.

**Batch:**

A `batch` is a subset of the total training data, i.e., a smaller portion of the entire training dataset. 

Instead of processing millions of samples at once, the training dataset is divided into smaller batches or mini-batches to manage memory and computational load for efficient model training. The model can process data in smaller chunks, avoiding problems like insufficient storage space and memory.

The batch determines how many samples will pass before updating the model's internal parameters (weights).

**Batch Size:**

A `batch size` indicates the number of training samples (such as the number of images or training data points) in a single batch (or mini-batch). 

For example, 1,000 data samples can be divided into ten batches of batch size of 100. The process of splitting the dataset is called `batch processing`.

Common batch sizes might choose the number of powers of 2, such as 32, 64, 128, 256, or 512.

**Iteration:**

A number that the subset of the training data will be processed based on a specific batch size.

**Note:**

Generally speaking, the model internal parameters (weights) are kept the same within 1 iteration (or 1 batch / 1 mini-batch). The model internal parameters will be updated after 1 iteration (or 1 batch).

## 2. Relationaship

The relationship among the `total number of samples`, `batch size`, and `iteration` is given by
$$ N_{total} = B \times I, $$
where 
$N_{total}$ is the `total number of samples`,
$B$ is the `batch size`, and
$I$ is the `number of iterations`.

**Note:**
Generally speaking, the model internal parameters (weights) are kept the same within 1 iteration (or 1 batch / 1 mini-batch). The model internal parameters will be updated after 1 iteration (or 1 batch).

**Example 1:**

For $N_{total}$ is the total number of samples of 1 epoch, the number of iterations is $I = N_{total}/B$.

We know that an epoch is composed of $I$ iterations ($N_{total}/B$ iterations) with the batch size $B$.

In this example, $N_{total} = N_{dataset}$, where $N_{dataset}$ is the number of samples of the entire dataset (or say "one complete pass of the entire training dataset").

**Example 2:**

For a training dataset of 1,000 data points, 1 epoch (one complete pass of the entire training dataset) contains the 1,000 training data points.

If a batch size of 200 is used, the number of iterations will be 5 (1000/200=5).

**Example 3:**

For a training dataset of 1,000 data points, 3 epoch contains the 3,000 training data points (1000*3=3000).

If a batch size of 200 is used, the number of iterations will be 15 (3000/200=15).

## 3. Simple description of machine learning in the 1st to $n$-th epoch

To understand the concept of an epoch in machine learning, let's briefly discuss a specific example of mini-batch processing.

**Simple example:** Training a model to predict game ratings (or a specific dataset). 

The entire dataset of 15,000 games (samples) with details, such as game category, game company, producer, and director. We could set the training to run for 25 epochs and choose a batch size of 150.

Now, here's how everything goes:

**`Dataset shuffle`, `batch processing`, `internal parameters (weights) update`, and `iteration`**

In the 1st epoch, the data is `shuffled` and then divided into 100 `batches` of 150 games each. 

$$ 150~\mathrm{(batch~size;~game/batch)} \times 100~\mathrm{(number~of~batches~or~iterations)} = 15000~\mathrm{(game)} $$

The model takes each batch (or iteration), analyzes the features of the 150 games, compares its predictions to actual ratings, and adjusts its internal parameters (weights). The model internal parameters (weights) are `updated` 100 times across all 100 batches (or iterations).

Then, data scientists/researchers `shuffle` the data, `batch` it again, and `iterate` the training process for the remaining 24 epochs. Eventually, the model has experienced the entire dataset 25 times (completing 25 epochs). During this process, its internal parameters (weights) were `updated` a total of 2500 times.

$$ 100~\mathrm{(batch/epoch)} \times 25~\mathrm{(epoch)} = 2500~\mathrm{(batch)} $$


## 4. Model internal parameters updating rate in 1 epoch

Model internal parameters (weights) are updated when a batch (or an iteration) is processed. There are three modes of `batch processing`:

**Full batch mode:**

The entire training dataset is treated as a single batch (i.e., $B = N_{dataset}$). This indicates that the model processes the entire dataset at once before updating its internal parameters (weights).

One major advantage of full batch mode processing is that the model can reach convergence rapidly (optimal performance) in fewer epochs than other modes.

On the other hand, full batch mode processing is computationally expensive, especially for complex models. It requires a lot amount of memory to hold the entire dataset. Unlike in other modes, there are no frequent parameter updates.

**Mini-batch mode:**

This is the most common type of batch processing mode. The training data is divided into smaller, manageable subsets/groups called mini-batches (or batches) ($1 < B < N_{dataset}$). The model trains one mini-batch at a time and updates its parameters after each mini-batch.

With the mini-batch mode, we could choose an appropriate batch size to avoid memory usage problems, meaning more speed and efficient. More frequent parameter updates is another advantage. However, it will require more epochs for convergence. We may also need to experiment a few times to find the optimal mini-batch size.

**Stochastic mode:**

The stochastic mode uses unique batch sizing ($B = 1$), which means the gradient and internal parameters (weights) are changed after each sample (instead of after each batch).

## 5. Advantages and disadvantages of various batch processing mode

The choice of batch size involves trade-offs between speed (efficiency), memory, and model accuracy: 

| Mode       | Batch size | Advantages | Disadvantages |
| ---------- | ---------- | ---------- | ------------- |
| Stochastic | $B = 1$ (one sample) | Low memory usage; can escape local minima. | Noisy updates; slower to complete an epoch. |
| Mini-batch | $1 < B < N_{dataset}$ | Balanced stability and speed; works well with GPUs. |Requires hyperparameter tuning. |
| Full batch | $B = N_{dataset}$ (all samples) | Stable, accurate gradient estimates. | High memory usage; prone to getting stuck in local minima. |

## 6. The importance of `epoch` in machine learning

`Epochs` are essential and important when training machine learning models. An `epoch` helps scientists/researchers to identify the model convergence (i.e., that best fits the training data). Here are a few more reasons why `epoch` is a crucial concept in machine learning.

**Model performance improvement:**  
The number of training `epochs` constitutes a critical hyperparameter in model optimization. An insufficient number of `epochs` may result in `underfitting`, whereby the model fails to capture the underlying structure of the data. Conversely, an excessive number of epochs increases the risk of `overfitting`, in which the model becomes overly adapted to noise and idiosyncrasies in the training dataset, thereby degrading its generalization capability on unseen data. Determining an appropriate number of `epochs` enables the model to effectively learn salient patterns while minimizing memorization of noise.

**Early stopping:**  
Continuous monitoring of performance metrics across `epochs` is essential for implementing `early stopping` strategies. `Early stopping` terminates the training process when performance on a validation set ceases to improve or begins to deteriorate. This approach not only mitigates `overfitting` but also enhances computational efficiency by avoiding unnecessary training iterations (saving computational resources).

**Better training dynamics:**  
Progression through training `epochs` provides valuable insights into the model's learning dynamics. By analyzing the rate of performance improvement and identifying points of convergence or plateau, scientists/researchers can make informed adjustments to hyperparameters, such as the learning rate, batch size, and model architecture. Such observations facilitate more effective optimization and improved overall model performance.

## 7. The difference between `epoch` and `batch`

As discussed above, a `batch` is a subset of the complete training dataset with a `batch size` of $1 \leq B \leq N_{dataset}$. Meanwhile, an `epoch` is when all batches complete one pass of the dataset through the algorithm.

A brief comparison between `epoch` and `batch` is listed below:

|        | Epoch        | Batch        |
| ------ | ------------ | ------------ |
| Concept | An epoch indicates all training dataset has completely passed through the training algorithm once. | A batch indicates all training data has been divided into smaller subsets for easier computational management. |
| Purpose | Epoch in machine learning does not directly address computational efficiency and is particularly about the learning regimen. | Batches are used to manage computational resources, ensuring models could be trained efficiently. |
| View in training | A macro view of the training process. | A micro view. |
| Value range | $1 \sim \infty$. | $1 \leq B \leq N_{dataset}$ |
| Number of samples per unit | $N_{dataset}$ per epoch | $1 \leq B \leq N_{dataset}$ per batch |
| Who sets | The number of epochs is a hyperparameter that the data scientists/researchers set to the training algorithm. | The batch size is also a hyperparameter that the data scientists/researchers sets. |
